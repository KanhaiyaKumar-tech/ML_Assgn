```python
a. To find fX(a) and fY(b), we need to integrate the joint PDF fxy(x, y) with respect to the respective variables.

fX(a) = ∫[0 to a] 6xy dy
      = 6∫[0 to a] xy dy
      = 6[a * ∫[0 to a] y dy]
      = 6[a * (y^2/2)][0 to a]
      = 6[a * (a^2/2)]
      = 3a^3, 0 ≤ a ≤ 1

fY(b) = ∫[0 to b] 6xy dx
      = 6∫[0 to b] xy dx
      = 6[y * ∫[0 to b] x dx]
      = 6[y * (x^2/2)][0 to b]
      = 6[y * (b^2/2)]
      = 3b^3, 0 ≤ b ≤ 1

b. To determine if A and B are independent, we need to check if the joint PDF fxy(x, y) can be factored into the product of the marginal PDFs fX(x) and fY(y).

fxy(x, y) = 6xy
fX(x) = ∫[0 to √x] 6xy dy
      = 3x^(3/2), 0 ≤ x ≤ 1

fY(y) = ∫[y^2 to 1] 6xy dx
      = 3(1 - y^4), 0 ≤ y ≤ 1

fX(x) * fY(y) = (3x^(3/2)) * (3(1 - y^4))
             = 9x^(3/2)(1 - y^4)

Since fxy(x, y) ≠ fX(x) * fY(y), A and B are not independent.

c. To find the conditional PDF of A given B = b, fA|B(a|b), we can use the formula:

fA|B(a|b) = fAB(a, b) / fB(b)

Since A and B are not independent, we can use the joint PDF fxy(x, y) to calculate fA|B(a|b):

fA|B(a|b) = fxy(a, b) / fY(b)
          = (6ab) / (3b^3)
          = 2a / b^2, 0 ≤ a ≤ 1, 0 ≤ b ≤ 1

d. To find E[A|B=b], we can use the conditional PDF fA|B(a|b):

E[A|B=b] = ∫[0 to 1] a * fA|B(a|b) da
         = ∫[0 to 1] a * (2a / b^2) da
         = (2 / b^2) ∫[0 to 1] a^2 da
         = (2 / b^2) * (a^3/3) [0 to 1]
         = 2 / (3b^2), 0 ≤ b ≤ 1

e. To find Var(A|B=b), we can use the conditional PDF fA|B(a|b) and the formula for variance:

Var(A|B=b) = E[(A - E[A|B=b])^2 | B=b]

Substituting E[A|B=b] from part d:

Var(A|B=b) = E[(A - 2 / (3b^2))^2 | B=b] 

```

8.There are 100 men on a ship. If Xi is the ith man&#39;s weight on the ship and Xi&#39;s are independent and
identically distributed and EXi = μ = 170 and σXi = σ = 30. Find the probability that the men&#39;s total
weight on the ship exceeds 18,000.

```python
To find the probability that the men's total weight on the ship exceeds 18,000, we need to calculate the probability distribution of the sum of their weights and then determine the probability that the sum exceeds 18,000.

Since the weights Xi are independent and identically distributed, the sum of their weights, denoted as S, follows a normal distribution. The mean of the sum is the sum of the means of the individual weights, and the standard deviation of the sum is the square root of the sum of the variances of the individual weights.

Let's denote the sum of the weights as S:

S = X1 + X2 + X3 + ... + X100

The mean of the sum is:

E[S] = E[X1 + X2 + X3 + ... + X100]
     = E[X1] + E[X2] + E[X3] + ... + E[X100]
     = 100 * μ
     = 100 * 170
     = 17,000

The standard deviation of the sum is:

σ[S] = sqrt(Var[X1 + X2 + X3 + ... + X100])
     = sqrt(Var[X1] + Var[X2] + Var[X3] + ... + Var[X100])
     = sqrt(100 * σ^2)
     = sqrt(100 * 30^2)
     = sqrt(90000)
     = 300

Now, we can calculate the probability that the men's total weight on the ship exceeds 18,000:

P(S > 18,000) = P((S - E[S]) / σ[S] > (18,000 - E[S]) / σ[S])
              = P(Z > (18,000 - 17,000) / 300)
              = P(Z > 1/3)

Here, Z represents a standard normal random variable. We can use a standard normal distribution table or a calculator to find the corresponding probability. For Z > 1/3, the probability is approximately 0.3707.

Therefore, the probability that the men's total weight on the ship exceeds 18,000 is approximately 0.3707 or 37.07%.
```
